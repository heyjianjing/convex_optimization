#### Intuition of duality

##### Lower bound of simple LP

We consider a simple LP

$$\min_{x, y} px+qy, \,\,\text{s.t. } x+y\geq 2, x\geq 0, y\geq 0$$

we want to know what is the lower bound of $px+qy$ for any given $p, q$

To approach this, we can consider a set of non-negative $a, b, c$, and we have

$$ax+ay\geq 2a, bx\geq 0, cy\geq 0$$

Add them up and regroup per $x, y$

$$(a+b)x+(a+c)y\geq 2a$$

Therefore, we can claim that, if we can find any set of $a, b, c$ such that they are `feasible` w.r.t.

$$a+b=p, a+c=q, a, b, c\geq 0$$

then, we get `a` lower bound of $px+qy$ that is $2a$

What about the best lower bound?

We can formulate it as

$$\max_{a,b,c} 2a,\,\,\text{s.t. } a+b=p, a+c=q, a,b,c\geq0$$

We see that the solution to this optimization provides the best `lower bound` to the `solution of original problem`, and we call this optimization the `dual` to the original `primal` problem

The number of `variables` in the dual is the number of `constraints` in the primal

If the dual is `infeasible`, by convention, its solution is maximizing over an empty set and is $-\infty$, and the primal would likely be unbounded below (e.g., the case where $p<0$)

(compared to the $+\infty$ convention for the primal when it is infeasible, and the dual would likely be unbounded above)

##### Lower bound of (still) simple LP

Now a simple LP with an equality constraint

$$\min_{x, y} px+qy, \,\,\text{s.t. } x\geq 0, y\leq 1, 3x+y=2$$

Similarly, for each constrain in the `primal`, we assign a `dual` variable $a\geq 0, b\geq 0, c$, and we have

$$ax\geq 0, -by\geq -b, (3x+y-2)c=0$$

We can now regroup them per primal variable $x, y$ to get

$$(a+3c)x+(-b+c)y\geq -b+2c$$

Therefore, we can formulate the dual problem

$$\max_{a,b,c} -b+2c,\,\,\text{s.t. }a+3c=p, -b+c=q, a,b\geq0$$

##### Lower bound of general LP

With $x\in \mathbf{R}^n$, given $c\in \mathbf{R}^n, A\in \mathbf{R}^{m \times n}, b\in \mathbf{R}^m, G\in \mathbf{R}^{p \times n}, h\in \mathbf{R}^p$, we can also consider a more general LP

$$\min c^Tx, \,\, \text{s.t. } Ax=b, Gx\leq h$$

For any feasible $x$, and for $\nu \in \mathbf{R}^m, \lambda \in \mathbf{R}^p, \lambda \geq 0$, we must have (since $Ax=b$, $Gx\leq h$)

$$\lambda^T(Gx-h)+\nu^T(Ax-b)\leq 0$$

Regroup, we have

$$(-A^T\nu-G^T\lambda)^Tx\geq -b^T\nu-h^T\lambda$$

If we let $-A^T\nu-G^T\lambda=c$, and we can formulate the following dual problem

$$\max_{\lambda, \nu} -b^T\nu-h^T\lambda,\,\, \text{s.t. } -A^T\nu-G^T\lambda=c, \lambda\geq 0$$

which generalizes the simple LP examples earlier

#### Lagrangian

From a different, but related perspective, for a standard (`convex or nonconvex`) optimization problem

$$\begin{align*}
\text{minimize}\,\,\, & f_0(x) \\
\text{subject to}\,\,\, & f_i(x)\leq 0, i=1, \cdots, m\\
& h_i(x)=0, i=1,\cdots, p
\end{align*}$$

with $x\in \mathbf{R}^n$, domain

$$D=\bigcap_{i=0}^m \text{dom }f_i \cap \bigcap_{i=1}^p\text{dom }h_i$$

and optimal value $p^*$ (primal), the `Lagrangian` is defined as

$$L(x, \lambda, \nu)=f_0(x)+\sum_{i=1}^m\lambda_if_i(x)+\sum_{i=1}^p\nu_ih_i(x)$$

where $\lambda_i$ and $\nu_i$ are Lagrange `multipliers` associated with the constraints (i.e., dual variables)

#### Lagrange `dual function`

The Lagrange dual function is defined as

$$g(\lambda, \nu)=\inf_{x\in D}L(x, \lambda, \nu)=\inf_{x\in D}\left(f_0(x)+\sum_{i=1}^m\lambda_if_i(x)+\sum_{i=1}^p\nu_ih_i(x)\right)$$

As $g$ is infimum of a family of affine functions in $\lambda, \nu$, it is `concave`

#### `Domain` of dual function

Unlike explicit domain defined for the primal problem, the domain of the dual function can be considered as defined by the set of $(\lambda, \nu)$ for which the dual function yields a finite value

$$\boxed{\text{dom } g=\{(\lambda, \nu)|\lambda\geq 0, \inf_{x\in D} L(x,\lambda,\nu)>-\infty\}}$$

#### `Lower bound property`

For any feasible $x$ and $\boxed{\lambda\geq 0}$ , we have (since $f_{1:m}(x)\leq 0$ and $h_i(x)=0$)

$$\boxed{f_0(x)\geq L(x, \lambda, \nu)\geq \inf_{x\in D}L(x, \lambda, \nu)=g(\lambda, \nu)}$$

Therefore, for $x^*$ such that $f_0(x^*)=p^*$, we have $p^*\geq g(\lambda, \nu)$

That is, the Lagrange dual function provides the `lower bound` for optimal value of the original problem

##### `Least norm` example

$$\begin{align*}
\text{minimize}\,\,\, & x^Tx \\
\text{subject to}\,\,\, & Ax=b
\end{align*}$$

We first write the Lagrangian

$$L(x, \nu)=x^Tx+\nu^T(Ax-b)$$

To get the dual function, we `minimize` w.r.t. $x$

$$\nabla_x L(x, \nu) = 2x + A^T\nu=0 \Longrightarrow x=-\frac{1}{2}A^T\nu$$

We plug this $x$ back to $L(x, \nu)$ to get $g(\nu)$

$$g(\nu) = \frac{1}{4}\nu^TAA^T\nu-\frac{1}{2}\nu^TAA^Tv-b^T\nu=-\frac{1}{4}\nu^TAA^Tv-b^T\nu$$

Therefore

$$p^*\geq -\frac{1}{4}\nu^TAA^Tv-b^T\nu, \forall \nu$$

##### `LP` example

$$\begin{align*}
\text{minimize}\,\,\, & c^Tx \\
\text{subject to}\,\,\, & -x\leq 0  \\
& Ax=b
\end{align*}$$

We first write the Lagrangian

$$\begin{align*}L(x, \lambda, \nu) &= c^Tx - \lambda^Tx +\nu^T(Ax-b) \\
& =-b^T\nu+(A^T\nu -\lambda+c)^Tx
\end{align*}$$

Since it is affine w.r.t. $x$, we have

$$g(\lambda, \nu)=\inf_{x} L(x, \lambda, \nu)=\left\{\begin{array}{ll}-b^T\nu & A^T\nu -\lambda+c=0 \\ -\infty & \text{otherwise} \end{array}\right.$$

Therefore, the lower bound property is

$$p^*\geq - b^T\nu, \text{if } A^T\nu+c\geq 0$$

##### Equality constrained `norm minimization` example

$$\begin{align*}
\text{minimize}\,\,\, & \|x\| \\
\text{subject to}\,\,\, & Ax=b
\end{align*}$$

Here, $\|\cdot\|$ is an `arbitrary norm`

We write the Lagrange dual function (note $Ax=b$ is treated as $b-Ax=0$)

$$g(\nu)=\inf_{x}(\|x\|-(A^T\nu)^Tx+b^T\nu)$$

If $\|A^Tv\|_*\leq 1$, from HÃ¶lder's inequality we have

$$(A^T\nu)^Tx\leq \|A^T\nu\|_*\|x\|\leq \|x\|$$

and the `equality` holds when $x=0$

Thus, for the dual function we have

$$g(\nu)=\inf_{x}(\|x\|-(A^T\nu)^Tx+b^T\nu)=0+b^T\nu = b^T\nu$$

If $\|A^Tv\|_*> 1$, then by definition of dual norm, $\exists x, \|x\|\leq 1, (A^Tv)^Tx=\|A^Tv\|_*>1$ and we have $\|x\|-(A^Tv)^Tx<0$

Now, if we scale $x$ to infinity through $kx, k\rightarrow \infty$, we see that $k\left(\|x\|-(A^Tv)^Tx\right) \rightarrow -\infty$

Put two scenarios together, we have

$$g(\nu)=\inf_{x}(\|x\|-(A^T\nu)^Tx+b^T\nu)=\left\{\begin{array}{ll}b^T\nu & \|A^Tv\|_*\leq 1 \\ -\infty & \|A^Tv\|_*> 1\end{array}\right.$$

##### Two way `partitioning` example

$$\begin{align*}
\text{minimize}\,\,\, & x^TWx \\
\text{subject to}\,\,\, & x_i^2=1, i=1, \cdots, n
\end{align*}$$

We write out the objective to see what it does

$$\begin{align*}
x^TWx & = \sum_{i=1}^nW_{ii}+2\sum_{i>j}W_{ij}x_ix_j
\end{align*}$$

Since only the second term contains $x_i$, the problem can be considered as having cost of
* $-W_{ij}$ as assigning $x_i, x_j$ to different signs
* $W_{ij}$ as assigning $x_i, x_j$ to the same sign

We write out the Lagrange dual function as

$$\begin{align*}
g(\nu) &= \inf_x\left(x^TWx + \sum_{i=1}^n \nu_i(x_i^2-1)\right) \\
& = \inf_x\left(x^T(W+\text{diag}(\nu))x\right)-1^T\nu \\
& = \left\{\begin{array}{ll}-1^T\nu & W+\text{diag}(\nu)\geq 0 \\ -\infty & \text{otherwise}\end{array}\right.
\end{align*}$$

This is because a quadratic form is lower bounded by

$$\lambda_{min} \|x\|_2^2 \leq x^TAx $$

Then, we have $A\geq 0\Rightarrow \lambda_{min} \geq 0 \Rightarrow x^TAx \geq 0$

However, if $\lambda_{min}<0$, then by letting $\|x\|_2 \rightarrow \infty$, $x^TAx \rightarrow -\infty$

##### `Analytic centering`

$$\begin{align*}
\text{minimize}\,\,\, & -\sum_{i=1}^n \log x_i\\
\text{subject to}\,\,\, & Ax=b, -x_i<0
\end{align*}$$

Write out the Lagrangian

$$\begin{align*}
L(x, \lambda, \nu)&=-\sum_{i=1}^n \log x_i + \sum_{i=1}^n\lambda_i (-x_i)+\nu^T(Ax-b) \\
&=-\sum_{i=1}^n \log x_i - \sum_{i=1}^n\lambda_i x_i+\nu^TAx-\nu^Tb \\
& A_i \text{ as ith column of }A\\
&=\sum_{i=1}^n \left(-\log x_i - \lambda_i x_i+(\nu^TA_i)x_i\right)-\nu^Tb \\
\end{align*}$$

Minimize w.r.t. $x_i$, we have

$$-\frac{1}{x_i}+\nu^TA_i-\lambda_i=0 \Longrightarrow x_i=\frac{1}{\nu^TA_i-\lambda_i}$$

Plug in this $x_i$ and we have the Lagrange dual function

$$\begin{align*}
g(\lambda, \nu)&=\inf_{x>0} L(x, \lambda, \nu)\\
& = \left\{\begin{array}{ll}\sum_{i=1}^n \log(\nu^TA_i-\lambda_i)-\nu^Tb+n & \nu^TA_i-\lambda_i>0 \\ -\infty & \text{otherwise}\end{array}\right.
\end{align*}$$

#### Lagrange dual function and `conjugate function`

For optimization problem

$$\begin{align*}
\text{minimize}\,\,\, & f_0(x) \\
\text{subject to}\,\,\, & Ax\leq b \\
& Cx=d
\end{align*}$$

We write out the Lagrange dual function and rearrange terms

$$\begin{align*}
g(\lambda, \nu)&=\inf_{x\in D}\left(f_0(x) - (-A^T\lambda - C^T \nu)^Tx\right)-b^T\lambda-d^T\nu \\
&=-\sup_{x\in D}\left( (-A^T\lambda - C^T \nu)^Tx-f_0(x) \right)-b^T\lambda-d^T\nu \\
& = -f_0^*(-A^T\lambda - C^T \nu)-b^T\lambda - d^T\nu
\end{align*}$$

Therefore, if we know the conjugate of $f_0$, we can write Lagrange dual function

#### `Dual feasibility`

$\lambda, \nu$ are `dual feasible` if
* $\lambda\geq 0$
* $(\lambda, \nu)\in \text{dom }g$

Once we have a point $\lambda, \nu$ that is dual feasible, we immediately get a `lower bound` for the original optimization problem

#### `Lagrange dual problem`

The Lagrange dual problem

$$\boxed{\begin{align*}
\text{maximize}\,\,\, & g(\lambda, \nu) \\
\text{subject to}\,\,\, & \lambda \geq 0
\end{align*}}$$

finds the `best lower bound` for $p^*$, which is the optimal value of the original `primal` problem

This is always a `convex optimization` problem since $g$ is concave, and its optimal value is often denoted $d^*$ (dual)

#### Implicit and explicit constraints

Implicit constraint $(\lambda, \nu)\in \text{dom } g$ is often made explicit

For example, `LP` primal problem

$$\begin{align*}
\text{minimize}\,\,\, & c^Tx \\
\text{subject to}\,\,\, & -x\leq 0  \\
& Ax=b
\end{align*}$$

and its dual (which is also an LP)

$$\begin{align*}
\text{maximize}\,\,\, & - b^T\nu \\
\text{subject to}\,\,\, & A^T\nu+c\geq 0
\end{align*}$$

The implicit constraint (associated with $\text{dom }g$) is $A^Tv-\lambda+c=0$. Since $\lambda \geq 0$, it becomes $A^T\nu+c\geq0$

#### Weak and strong duality

`Weak` duality: $d^* \leq p^*$, can be used to find nontrivial lower bounds for difficult problems

For example, the dual problem for two-way partitioning can be solved by SDP

$$\begin{align*}
\text{maximize}\,\,\, & - 1^T\nu \\
\text{subject to}\,\,\, & W+\text{diag}(\nu)\geq 0
\end{align*}$$

`Strong` duality: $d^* = p^*$

(Often) holds for convex problems

`Sufficient conditions` that guarantee strong duality are called `constraint qualifications`

#### `Slater's` constraint qualification

Strong duality holds for a `convex` problem

$$\begin{align*}
\text{minimize}\,\,\, & f_0(x) \\
\text{subject to}\,\,\, & f_i(x)\leq 0, i=1, \cdots, m\\
& Ax=b
\end{align*}$$

if the problem is `strictly` feasible

$$\exists x\in \text{int } D, f_i(x)<0, i=1, \cdots, m$$

#### `Inequality form` LP and its dual

$$\begin{align*}
\text{minimize}\,\,\, & c^Tx \\
\text{subject to}\,\,\, &  Ax\leq b
\end{align*}$$

We write out the dual function

$$\begin{align*}g(\lambda)&=\inf_x(c^Tx+\lambda^T(Ax-b))\\
&= \inf_x (c+A^T\lambda)^Tx-b^T\lambda \\
&=\left\{\begin{array}{ll}-b^T\lambda & c+A^T\lambda =0 \\ -\infty & \text{otherwise}\end{array}\right.
\end{align*}$$

Therefore, we have the dual problem

$$\begin{align*}
\text{maximize}\,\,\, & -b^T\lambda \\
\text{subject to}\,\,\, &  c+A^T\lambda =0 \\
&\lambda\geq 0
\end{align*}$$

Slater's condition says that $d^*=p^*$ if $Ax<b$ for some $x$, which basically says that the polyhedron has `nonempty interior`

#### `Inequality form` QP and its dual

$$\begin{align*}
\text{minimize}\,\,\, & x^TPx \\
\text{subject to}\,\,\, &  Ax\leq b
\end{align*}$$

We write out the dual function (for $P\in S^n_{++}$)

$$\begin{align*}g(\lambda) &=\inf_x (x^TPx+\lambda^T(Ax-b)) \\
& = \inf_x(x^TPx+\lambda^TAx)-b^T\lambda\\
& \text{take derivative and set to zero for } x^TPx+\lambda^TAx\\
& \Rightarrow x=-\frac{1}{2}P^{-1}q, \,q=A^T\lambda  \\
&= -\frac{1}{4}q^TP^{-1}q-b^T\lambda \\
& = -\frac{1}{4}\lambda^TAP^{-1}A^T\lambda-b^T\lambda
\end{align*}$$

Therefore, the dual problem is

$$\begin{align*}
\text{maximize}\,\,\, & -\frac{1}{4}\lambda^TAP^{-1}A^T\lambda-b^T\lambda \\
\text{subject to}\,\,\, &  \lambda\geq 0
\end{align*}$$

#### Complementary slackness

If `strong duality` holds for an optimization problem, and $f_0(x^*)=p^*$, $g(\lambda^*, \nu^*)=d^*$, then, we have

$$\begin{align*}
f_0(x^*)&=g(\lambda^*, \nu^*) \\
& \text{definition of Lagrange dual function} \\
& = \inf_{x\in D}\left(f_0(x)+\sum_{i=1}^m\lambda_i^*f_i(x)+\sum_{i=1}^p\nu_i^*h_i(x)\right) \\
& \text{property of infimum} \\
& \leq f_0(x^*)+\sum_{i=1}^m\lambda_i^*f_i(x^*)+\sum_{i=1}^p\nu_i^*h_i(x^*) \\
& x^* \text{ is feasible}, \lambda^*\geq 0 \Rightarrow \sum_{i=1}^m\lambda_i^*f_i(x^*)\leq 0, \sum_{i=1}^p\nu_i^*h_i(x^*)=0 \\
&\leq f_0(x^*)
\end{align*}$$

Therefore, all inequalities become `equalities` and we have
* $x^*$ `minimizes the Lagrangian` $L(x, \lambda^*, \nu^*)$ since

$$\begin{align*}\inf_{x\in D}L(x, \lambda^*, \nu^*)&=\inf_{x\in D}\left(f_0(x)+\sum_{i=1}^m\lambda_i^*f_i(x)+\sum_{i=1}^p\nu_i^*h_i(x)\right) \\
&= f_0(x^*)+\sum_{i=1}^m\lambda_i^*f_i(x^*)+\sum_{i=1}^p\nu_i^*h_i(x^*) \\
& = L(x^*, \lambda^*, \nu^*)
\end{align*}$$
* Since $\sum_{i=1}^p\nu_i^*h_i(x^*)=0$, $\sum_{i=1}^m\lambda_i^*f_i(x^*)$ must be zero as well
* Since $\sum_{i=1}^m\lambda_i^*f_i(x^*)$ is sum of product between nonnegative and nonpositive numbers, we have
$$\boxed{\lambda_i^*f_i(x^*)=0, i=1, \cdots, m}$$
This is known as `complementary slackness`
* As a result, for the `ith` constraint
$$\lambda_i^*>0\Rightarrow f_i(x^*)=0, \,f_i(x^*)<0 \Rightarrow \lambda_i^*=0$$

#### Karush-Kuhn-Tucker (KKT) conditions

If `strong duality` holds (convex or not), $x, (\lambda, \nu)$ are primal and dual `optimal` pair, $f_i, h_i$ are `differentiable`, then they must satisfy the following KKT conditions (`necessary` conditions)

* Primal feasibility
$$f_i(x)\leq 0, i=1,\cdots, m, \, h_i(x)=0, i=1, \cdots, p$$
* Dual feasibility
$$\lambda\geq 0$$
* Complementary slackness
$$\lambda_i^*f_i(x^*)=0, i=1, \cdots, m$$
* Gradient of Lagrangian w.r.t. $x$ vanishes (stationary)
$$\nabla f_0(x)+\sum_{i=1}^m \lambda_i \nabla f_i(x)+\sum_{i=1}^p \nu_i \nabla h_i(x)=0$$

The gradient of the Lagrangian w.r.t. primal variable is commonly known as `dual residual`

$$\boxed{r_d = \nabla f_0(x)+\sum_{i=1}^m \lambda_i \nabla f_i(x)+\sum_{i=1}^p \nu_i \nabla h_i(x)}$$

Thus, the dual residual measures how far the current solution is from satisfying the stationary condition in KKT

#### KKT as `sufficient` conditions for `convex` problem

Suppose $\tilde{x}, \tilde{\lambda}, \tilde{\nu}$ satisfy KKT for a `convex` problem (so $f_i$ and $h_i$ must be differentiable)

$$\begin{align*}
\text{minimize}\,\,\, & f_0(x) \\
\text{subject to}\,\,\, & f_i(x)\leq 0, i=1, \cdots, m\\
& Ax=b
\end{align*}$$

then, from `complementary slackness` $\tilde{\lambda}_if_i(\tilde{x})=0, i=1, \cdots, m$, we have

$$\begin{align*}L(\tilde{x}, \tilde{\lambda}, \tilde{\nu})&=f_0(\tilde{x})+\sum_{i=1}^m\tilde{\lambda}_if_i(\tilde{x})+\sum_{i=1}^p\tilde{\nu}_ih_i(\tilde{x}) \\
& =f_0(\tilde{x})
\end{align*}$$



In addition, with $\lambda\geq 0$, the Lagrangian of a convex problem is `convex` (w.r.t. $x$) and

$$g(\lambda, \nu)=\inf_{x\in D}L(x, \lambda, \nu)=\inf_{x\in D}\left(f_0(x)+\sum_{i=1}^m\lambda_if_i(x)+\sum_{i=1}^p\nu_ih_i(x)\right)$$

Therefore, $g(\lambda, \nu)$ gets optimal value if (from 1st condition for convexity)

$$\nabla_x L=\nabla f_0(x)+\sum_{i=1}^m \lambda_i \nabla f_i(x)+\sum_{i=1}^p \nu_i \nabla h_i(x)=0 $$

which is satisfied by $\tilde{x}, \tilde{\lambda}, \tilde{\nu}$ from the fourth condition in KKT, or

$$g(\tilde{\lambda}, \tilde{\nu}) = L(\tilde{x}, \tilde{\lambda}, \tilde{\nu})$$

Hence, $\tilde{x}, \tilde{\lambda}, \tilde{\nu}$ are optimal

$$f_0(\tilde{x})=g(\tilde{\lambda}, \tilde{\nu})$$

If `Slater's condition` is satisfied, then $x$ is optimal `if and only if` there exist $\lambda, \nu$ that satisfy KKT conditions

#### Example of convex problem

$$\begin{align*}
\text{minimize}\,\,\, & -\sum_{i=1}^n \log(x_i+\alpha_i), \alpha_i>0\\
\text{subject to}\,\,\, & -x\leq 0\\
& 1^Tx=1
\end{align*}$$

We write out the Lagrangian

$$L(x, \lambda, \nu)=-\sum_{i=1}^n \log(x_i+\alpha_i)-\lambda^T x + \nu(1^Tx-1)$$

We can let $x_i=\frac{1}{n}$ to show that Slater's condition holds, therefore, $x$ is optimal `if and only if` there exist $\lambda\in \mathbf{R}^n, v\in \mathbf{R}$ such that KKT is satisfied
* $x\geq 0, 1^Tx=1$
* $\lambda\geq 0$
* $\lambda_i x_i=0, i=1, \cdots, n$
* $\nabla_x L=0 \Rightarrow \frac{1}{x_i+\alpha_i}+\lambda_i=\nu, i=1, \cdots, n \Rightarrow \nu \geq \frac{1}{x_i+\alpha_i}$

If $\nu< 1/\alpha_i$, then the only way that #4 holds is $x_i>0$, which means $\lambda_i=0$

If $\nu\geq 1/\alpha_i$, then the only way that #4 holds is $\lambda_i>0$, which means $x_i=0$

Either case, we get rid of either $\lambda_i$ or $x_i$ and can solve for $\nu$

#### `Perturbed` primal and dual

If we `perturb` the original problem a bit

$$\begin{align*}
\text{minimize}\,\,\, & f_0(x) \\
\text{subject to}\,\,\, & f_i(x)\leq u_i, i=1, \cdots, m\\
& h_i(x)=v_i, i=1,\cdots, p
\end{align*}$$

then, the optimal value $p^*(u, v)$ becomes a function of $u, v$, with $p^*(0, 0)$ the optimal value of the unperturbed problem, and its dual problem becomes

$$\begin{align*}
\text{maximize}\,\,\, & g(\lambda, \nu)-u^T\lambda-v^T\nu \\
\text{subject to}\,\,\, & \lambda\geq 0
\end{align*}$$

#### Global sensitivity

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')
np.set_printoptions(formatter={'float': '{: 0.2e}'.format})
```

```python
def f(x):
    return np.exp(-(x-1))-1

def grad_f(x):
    return -np.exp(-(x-1))

def lower_bound(x, x_0):
    return f(x_0) + grad_f(x_0) * (x - x_0)

x_0 = 0
x = np.linspace(-1, 4, 400)

plt.plot(x, f(x), label="$p^*(u)$", color='w', linewidth=2)
plt.plot(x, lower_bound(x, x_0), label="$p^*(0)-\lambda^*u$", linestyle='--', color='c')
plt.scatter(x_0, f(x_0), color='w', zorder=5)
plt.text(0.5, 1.7, '$p^*(0)$', horizontalalignment='center', fontsize=12, color='w')
plt.axhline(0, color='w', ls='--', linewidth=1)
plt.axvline(0, color='w', ls='--', linewidth=1)

plt.ylim(-2, 6)
plt.xlim(-1, 4)
plt.xlabel("$u$")
plt.legend()
plt.show()
```

Suppose `strong duality` holds for unperturbed problem with $\lambda^*,\nu^*$ being dual optimal, and assume `weak duality` to perturbed problem and we have `global` sensitivity result

$$\begin{align*}p^*(u, v) &\geq g(\lambda^*, \nu^*)-u^T\lambda^*-v^T\nu^* \\
&=p^*(0,0)-u^T\lambda^*-v^T\nu^*
\end{align*}$$

However, due to being just a lower bound, we often cannot say much how the actual $p^*(u, v)$ changes as we loosen or tighten the constraints

In addition, the relationship between $p^*(u, v)$ and the perturbations can be complex when it is far from the unperturbed point $p^*(0,0)$

#### Local sensitivity and interpretation of Lagrange multipliers

We can show that if $p^*(u, v)$ is differentiable at $(0, 0)$, then

$$\lambda_i^*=\left.-\frac{\partial p^*(u, v)}{\partial u_i}\right|_{u=0, v=0}, \nu_i^*=\left.-\frac{\partial p^*(u, v)}{\partial v_i}\right|_{u=0, v=0}$$

These allow us to explore local behavior of optimal value when we change $u_i, v_i$

First, the scenario when the constraint is `slack`, for example, if $f_3(x^*)=-0.1$, then by complementary slackness, $\lambda^*_3=0$

This means that we can change $u_i$ locally and there will be no (first-order) effect on the optimal value $p^*$

Then, there is also the scenario when the constraint is `tight`, for example

$$f_1(x^*)=0, \lambda_1^*=0.001,\,\, f_2(x^*)=0, \lambda_2^*=10, \cdots$$

With these, we can say something when we have to redesign the $x^*$

For example, if we are asked to remove some resource (i.e., tighten the constraint and go negative), then we better start with $f_1$ due to smaller $\lambda^*_1$ to minimize impact on optimal value

And if we are asked to make the biggest difference with minimal adjustment of resource, then we may start with $f_2$
